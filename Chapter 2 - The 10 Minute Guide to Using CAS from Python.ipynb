# Chapter 2 - The 10 Minute Guide to Using CAS from Python

A quick introduction to the basics of using the SWAT package to connect to CAS and get action results.

# Importing SWAT and Getting Connected

Import the SWAT package and create a connection.

```python
import swat
```

```python
conn = swat.CAS('server-name.mycompany.com', 5570)
```

Run the **help** action.

```python
out = conn.help()
```

List the keys in the result.

```python
list(out.keys())
```

Display the contents of the result.

```python
out
```

Access the 'builtins' key of the result.

```python
out['builtins']
```

Access the 'builtins' key of the result using attribute syntax.

```python
out.builtins
```

## Running CAS Actions

Run the **userinfo** action.

```python
conn.userinfo()
```

You can use the **pprint** module to "pretty print" the result as a dictionary.

```python
from pprint import pprint

pprint(dict(conn.userinfo()))
```

Call the **userinfo** action using the fully-qualified action name.

```python
conn.builtins.userinfo()
```

## Loading Data

Using the **upload** method of a **CAS** connection object.

```python
out = conn.upload('https://raw.githubusercontent.com/' +
                          'pydata/pandas/master/pandas/tests/' + 
                          'data/iris.csv')
out
```

Grab the casTable key of the output.  It's a **CASTable** object that references the CAS table created on the server.

```python
iris = out.casTable
```

Run the **tableinfo** action on the **CASTable** object.

```python
iris.tableinfo()
```

Run the **columninfo** action on the **CASTable** object.

```python
iris.columninfo()
```

## Executing Actions on CAS Tables

Using IPython's help hooks to display help for the **simple** action set.

```python
conn.simple?
```

Run the **summary** action on the **CASTable**.

```python
summ = iris.summary()
summ
```

Run the **describe** method on the **CASTable** object.  This method emulates the Pandas **describe** method, but calls CAS actions to compute the results.

```python
iris.describe()
```

## Data Visualization

Using the **plot** method of the **CASTable**.

```python
iris.plot()
```

Plots may not display automatically depending on your configuration.  You can use the **show** method of the plot to display it.

```python
import matplotlib.pyplot as plt
```

```python
plt.show()
```

Alternatively, if you are in a Jupyter notebook, you can tell matplotlib to display images inline using the **%matplotlib** magic.

```python
%matplotlib inline
```

```python
iris.plot()
```

Sort the values in the table so that the order in the plot is predictable.

```python
iris.sort_values(['SepalLength', 'SepalWidth']).plot()
```

Close the connection.

```python
conn.close()
```
